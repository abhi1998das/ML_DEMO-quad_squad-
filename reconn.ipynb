```python
import pandas as pd
import numpy as np
import keras

```

```python
df = pd.read_csv('Data1.csv')

df.drop(columns=['StudentID','First Name','Last Name'],inplace=True)
```

```python
print(df.head(5))
```

```python
df['HTML']=df['HTML'].astype(int)
df['CSS']=df['CSS'].astype(int)
df['PHP']=df['PHP'].astype(int)
df['JAVASCRIPT']=df['JAVASCRIPT'].astype(int)
df['SQL']=df['SQL'].astype(int)
```

```python
x=df.iloc[:, 0:8]
y=df.iloc[:, 8]
```

```python
from sklearn import preprocessing
s1=preprocessing.LabelEncoder()
x.iloc[:, 0]= s1.fit_transform(x.iloc[:, 0])
s2=preprocessing.LabelEncoder()
x.iloc[:, 1]= s2.fit_transform(x.iloc[:, 1])
```

```python
s3=preprocessing.LabelEncoder()
y= s3.fit_transform(y)
```

```python
y = np.reshape(y, (-1, 1))
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
y=enc.fit_transform(y).toarray()
```

```python

s1=preprocessing.LabelEncoder()
x.iloc[:, 0]= s1.fit_transform(x.iloc[:, 0])
enc = OneHotEncoder(categorical_features=[1])
x=enc.fit_transform(x).toarray()
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense,Flatten,Dropout
```

```python
 model = Sequential([
            Dense(48, input_shape=(11,), activation='relu'),
            Dense(48, activation='relu'),
            Dense(10, activation='sigmoid')
            ])
```

```python
model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=40)
```

```python
print(x_test.shape)
```

```python
model.fit(x_train, y_train, epochs=1000, batch_size=32)
```

```python
y_pred=model.predict(x_test)
y_pred =(y_pred>=0.5)
y_pred = y_pred.astype(int)
print(y_pred)
```
